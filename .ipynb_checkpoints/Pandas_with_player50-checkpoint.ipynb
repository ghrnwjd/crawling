```python
# 칼럼 이름을 바꾸고 저장하기 
# team >> club

df.rename(columns = {'team' : 'club'}, inplace = True) # inplace == True 로 하면 저장 됌
df.head()
```

### 데이터 전처리(pre-processing)

```python
# 유로화 단위를 없애기

#df['value'] = df['value'].str.replace('m', '').astype('float')
print(df['value'].head())

df.info()
```

```python
df.head()
```

```python
# 유로 백만단위 -> 13을 곱하면 억
df['시장가치(억)'] = df['value']*13
df.head()
```

```python
# 칼럼 삭제 df.drop(columns = ['칼럼이름'])
df.drop(columns=['value'], inplace= True)
df.head()
```

### DataFrame 통계분석과 groupby()

```python
# 숫자형 데이터에 대한 통계
df.describe()
```

```python
# df[].mean << 숫자형에 관해서만
df['age'].mean()
```

```python
# 몸값 합계구하기
sum = df['시장가치(억)'].sum()
sum
```

```python
#선수들이 속한 가장 많은 나라 최빈값 mode()
df['nation'].mode()
```

```python
# 국적이 브라질인 선수는?
df[df['nation'] == 'Brazil']
```

```python
# groupby()

g = df.groupby('nation')
g.size()
g.count()
```

```python
# g.sum()
g['시장가치(억)'].sum()
```

```python
team = df.groupby('club')
team.sum()
team['시장가치(억)'].sum().sort_values(ascending=False)
```
